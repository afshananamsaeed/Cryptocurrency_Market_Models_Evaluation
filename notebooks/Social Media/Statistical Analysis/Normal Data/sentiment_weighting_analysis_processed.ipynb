```python
bitcoin_hourly = pd.read_csv("/mnt/bitcoin_hourly_prices_till_2023-03-30.csv", lineterminator='\n')
bitcoin_minutely = pd.read_csv("/mnt/bitcoin_price_minutely_data_till_2023_03_30.csv", lineterminator='\n')
```

```python
data_30min = pd.read_csv("/home/ubuntu/Masters Thesis/data/final/sentiment_weighted_scores/Without_bots/weighted_scores_bitcoin_tweets_wo_bots_interval_30T_processed.csv", lineterminator='\n')
data_1hrs = pd.read_csv("/home/ubuntu/Masters Thesis/data/final/sentiment_weighted_scores/Without_bots/weighted_scores_bitcoin_tweets_wo_bots_interval_1H_processed.csv", lineterminator='\n')
data_4hrs = pd.read_csv("/home/ubuntu/Masters Thesis/data/final/sentiment_weighted_scores/Without_bots/weighted_scores_bitcoin_tweets_wo_bots_interval_4H_processed.csv", lineterminator='\n')
data_6hrs = pd.read_csv("/home/ubuntu/Masters Thesis/data/final/sentiment_weighted_scores/Without_bots/weighted_scores_bitcoin_tweets_wo_bots_interval_6H_processed.csv", lineterminator='\n')
data_1day = pd.read_csv("/home/ubuntu/Masters Thesis/data/final/sentiment_weighted_scores/Without_bots/weighted_scores_bitcoin_tweets_wo_bots_interval_D_processed.csv", lineterminator='\n')
```

```python
data_30min = data_30min[~data_30min['spacytextblob_processed_data_polarity_score'].isna()]
data_1hrs = data_1hrs[~data_1hrs['spacytextblob_processed_data_polarity_score'].isna()]
data_4hrs = data_4hrs[~data_4hrs['spacytextblob_processed_data_polarity_score'].isna()]
data_6hrs = data_6hrs[~data_6hrs['spacytextblob_processed_data_polarity_score'].isna()]
data_1day = data_1day[~data_1day['spacytextblob_processed_data_polarity_score'].isna()]

data_30min = data_30min.drop(columns=['Unnamed: 0'])
data_1hrs = data_1hrs.drop(columns=['Unnamed: 0'])
data_4hrs = data_4hrs.drop(columns=['Unnamed: 0'])
data_6hrs = data_6hrs.drop(columns=['Unnamed: 0'])
data_1day = data_1day.drop(columns=['Unnamed: 0'])
```

```python
bitcoin_minutely = bitcoin_minutely.rename(columns={'time': 'date'})
bitcoin_minutely['date'] = bitcoin_minutely['date'].apply(datetime.datetime.utcfromtimestamp)
bitcoin_hourly = bitcoin_hourly.rename(columns={'datetimes': 'date'})
bitcoin_hourly['date'] = pd.to_datetime(bitcoin_hourly['date'])
```

```python
data_30min_joined = GetModellingData(data=data_30min, bitcoin_data=bitcoin_minutely, frequency='30T').get_data_for_feature_analysis()
data_1hr_joined = GetModellingData(data=data_1hrs, bitcoin_data=bitcoin_hourly, frequency='1H').get_data_for_feature_analysis()
data_4hr_joined = GetModellingData(data=data_4hrs, bitcoin_data=bitcoin_hourly, frequency='4H').get_data_for_feature_analysis()
data_6hr_joined = GetModellingData(data=data_6hrs, bitcoin_data=bitcoin_hourly, frequency='6H').get_data_for_feature_analysis()
data_1d_joined = GetModellingData(data=data_1day, bitcoin_data=bitcoin_hourly, frequency='1D').get_data_for_feature_analysis()
```

```python
data_30min_joined.replace([np.inf, -np.inf], np.nan, inplace=True)
data_30min_joined.dropna(inplace=True)
data_1hr_joined.replace([np.inf, -np.inf], np.nan, inplace=True)
data_1hr_joined.dropna(inplace=True)
data_4hr_joined.replace([np.inf, -np.inf], np.nan, inplace=True)
data_4hr_joined.dropna(inplace=True)
data_6hr_joined.replace([np.inf, -np.inf], np.nan, inplace=True)
data_6hr_joined.dropna(inplace=True)
data_1d_joined.replace([np.inf, -np.inf], np.nan, inplace=True)
data_1d_joined.dropna(inplace=True)
```

```python
# statistical tests - multiple linear regression
import statsmodels.api as sm

X = sm.add_constant(data_30min_joined[['flair_processed_data_score', 'spacytextblob_processed_data_polarity_score', 'spacytextblob_processed_data_subjectivity_score', 'textblob_processed_data_polarity_score', 'textblob_processed_data_subjectivity_score', 'vader_processed_data_score', 'bert_processed_data_sentiment_score', 'roberta_processed_data_sentiment_score']])

# on close bitcoin price for the given date
model = sm.OLS(data_30min_joined['close'], X).fit()
print(f"model summary for same day closing price{model.summary()}")

# on close bitcoin price for the given date
model = sm.OLS(data_30min_joined['close_t+1'], X).fit()
print(f"model summary for day+1 closing price{model.summary()}")

# on close bitcoin price for the given date
model = sm.OLS(data_30min_joined['close_t+2'], X).fit()
print(f"model summary for day+2 day closing price{model.summary()}")

# on close bitcoin price for the given date
model = sm.OLS(data_30min_joined['close_t+3'], X).fit()
print(f"model summary for day+3 day closing price{model.summary()}")

# on close bitcoin price for the given date
model = sm.OLS(data_30min_joined['close_t+7'], X).fit()
print(f"model summary for day+7 day closing price{model.summary()}")
```
