```python
import pandas as pd
```

```python
data_np = pd.read_csv('/mnt/model_output_nn/user_weighted/dataset1/parameter_tuning/output_manual_processed_False_SPT_5_5_1.csv')
data_p = pd.read_csv('/mnt/model_output_nn/user_weighted/dataset1/parameter_tuning/output_manual_processed_True_SPT_5_5_1.csv')
```

```python
data_grouped_np = data_np.groupby(['network_name', 'feature']).apply(lambda x: x.loc[x['test_mae'].idxmin()])
data_grouped_np.reset_index(drop=True)
data_grouped_np['processed'] = False
data_grouped_np['sentiment_past'] = 5
data_grouped_np['price_past'] = 5
data_grouped_np['forecast'] = 1
```

```python
data_grouped_p = data_p.groupby(['network_name', 'feature']).apply(lambda x: x.loc[x['test_mae'].idxmin()])
data_grouped_p.reset_index(drop=True, inplace=True)
data_grouped_p['processed'] = True
data_grouped_p['sentiment_past'] = 5
data_grouped_p['price_past'] = 5
data_grouped_p['forecast'] = 1
```

```python
stacked_df = pd.concat([data_grouped_np, data_grouped_p], axis=0)
stacked_df.reset_index(drop=True, inplace=True)
# stacked_df.to_csv()
```
