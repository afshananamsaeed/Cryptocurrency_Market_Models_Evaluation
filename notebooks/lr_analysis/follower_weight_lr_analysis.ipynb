```python
import pandas as pd
```

```python
data_np = pd.read_csv('/mnt/model_output_nn/user_weighted/dataset1/parameter_tuning/output_manual_processed_False_SPT_15_15_7_follower_weighted.csv')
data_p = pd.read_csv('/mnt/model_output_nn/user_weighted/dataset1/parameter_tuning/output_manual_processed_True_SPT_15_15_7_follower_weighted.csv')
```

```python
data_grouped_np = data_np.groupby(['network_name', 'feature']).apply(lambda x: x.loc[x['test_mae'].idxmin()])
data_grouped_np.reset_index(drop=True)
# data_grouped_np['processed'] = False
# data_grouped_np['sentiment_past'] = 15
# data_grouped_np['price_past'] = 15
# data_grouped_np['forecast'] = 7
```

```python
data_grouped_p = data_p.groupby(['network_name', 'feature']).apply(lambda x: x.loc[x['test_mae'].idxmin()])
data_grouped_p.reset_index(drop=True, inplace=True)
# data_grouped_p['processed'] = True
# data_grouped_p['sentiment_past'] = 15
# data_grouped_p['price_past'] = 15
# data_grouped_p['forecast'] = 7
```
